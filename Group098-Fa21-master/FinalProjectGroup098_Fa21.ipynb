### Why We Are Interested In This Question
Through prior knowledge and research, our group has found that food insecurity causes major negative implications for people of all ages, such as: health concerns, hindering child development, and  increased mental health issues [2] [4]. # Data Cleaning
## Food Access Research Atlas (Main Dataset)

Due to the larger scale of our dataset (looking at food across the U.S.), we decided to focus on the west coast, specifically California, to further narrow our research question. We decided it made sense to focus on California since it is considered one of the more ethnically and financially diverse states in the U.S. Other states may not have had enough data about more races to make any meaningful conclusions.
 
We are considering factors like geographic distance and race, and our dataset goes into extensive depth about each of these factors. See all the initial data collected by the USDA for the west coast below.

```python
# Only keep data from CA, WA, and OR (west coast states)
w_coast = food_access[(food_access['State']=='California') | 
    (food_access['State']=='Washington') | (food_access['State']=='Oregon')]

w_coast.head()
```

Our clean up included removing rows and columns with more than 60 percent null data. The data beyond the 60 percent null data threshold was too incomplete to be useful in our analysis. Because we are considering factors like race into our analysis, if we are missing too many variables, then we are unable to get a relevant and complete analysis, therefore the null data was only cluttering our dataset. By removing this portion, we tidied up our data, making sure that the data left was numerical (as opposed to strings like ‘NA’). By making sure the data we are analyzing is in a uniform numerical format, we can avoid programming errors later on during data analysis. 

```python
# Remove rows/cols with more than 60% null data
w_coast = w_coast.dropna(thresh=math.ceil(w_coast.shape[1]*0.4),axis=0)
w_coast = w_coast.dropna(thresh=math.ceil(w_coast.shape[0]*0.4),axis=1)

w_coast.head()
```

From the remaining columns, we filtered through the columns and checked the variables and variable descriptions to keep the columns we thought would be useful for our data analysis. In general, we kept columns decribing the state, county, populations, and populations of specific ethnicities/races within a mile because this would be relevant to our research question of whether factors like location and race affect extent of food insecurity. We also renamed the columns for easier readability.

```python
# Keep columns that are relevant to our research question
# We kept columns that had to do with location, population count, and state and county description.
cols_to_keep = ['State', 'County', 'Urban', 'Pop2010', 'OHU2010','LowIncomeTracts', 'PovertyRate', 'MedianFamilyIncome',
    'LA1and10', 'LAhalfand10', 'LA1and20', 'LATracts_half', 'LATracts1','LATracts10', 'LATracts20','LALOWI1_10','lapop1',
    'lalowi1','lawhite1', 'lablack1','laasian1','lanhopi1','laaian1','laomultir1','lahisp1','lasnap1','TractLOWI',
    'TractWhite', 'TractBlack', 'TractAsian', 'TractNHOPI', 'TractAIAN','TractOMultir', 'TractHispanic','TractSNAP']
df = w_coast[cols_to_keep]

# Low access = location is beyond supermarket
# Rename columns for easier readability
df.columns = ['State', 'County', 'Urban Flag','Population','Housing Units','Low Income Flag',
    'Poverty Rate', 'Median Family Income','1 mi(Urban), 10 mi(Rural) Low Access Flag',
    '0.5 mi(Urban), 10 mi(Rural) Low Access Flag','1 mi(Urban), 20 mi(Rural) Low Access Flag','0.5 mi Low Access Flag',
    '1 mi Low Access Flag', '10 mi Low Access Flag','20 mi Low Access Flag','1 mi (Urban), 10 mi(Rural) Low Income and Access Pop',
    '1 mi Low Access Pop','1 mi Low Access & Income Pop','1 mi Low Access White Pop','1 mi Low Access African American Pop',
    '1 mi Low Access Asian Pop','1 mi Low Access Pacific Islander Pop','1 mi Low Access Native Pop','1 mi Low Access Other/Mult. Race Pop',
    '1 mi Low Access Hispanic Pop','1 mi Low Access SNAP Pop','Low Income Pop','White Pop','African American Pop', 
    'Asian Pop','Pacific Islander Pop','Native Pop','Other/Mult. Race Pop','Hispanic Pop', 'SNAP housing unit Pop']
df.head()
```

Finally, we split the data to only include California in a separate dataframe for easier accessibility and readability. After this, we summed each of the population counts for each county since they were separated by census tracts.

```python
CA = df[df['State']=='California']

CA = CA.groupby('County').agg(


    Population=('Population',sum),


    Median_Income=('Median Family Income',sum),
    Low_Access_and_Income_Pop=('1 mi Low Access & Income Pop',sum),

    #calculating what proportion of tracts in a county have low access flag set for specific radiuses named
    Prop_Tracts_1_10_mi_Low_Access=('1 mi(Urban), 10 mi(Rural) Low Access Flag',np.mean),
    Prop_Tracts_half_10_mi_Low_Access=('0.5 mi(Urban), 10 mi(Rural) Low Access Flag',np.mean),
    Prop_Tracts_1_20_mi_Low_Access=('1 mi(Urban), 20 mi(Rural) Low Access Flag',np.mean),
    Prop_Tracts_half_mi_Low_Access=('0.5 mi Low Access Flag',np.mean),
    Prop_Tracts_1_mi_Low_Access=('1 mi Low Access Flag',np.mean),
    Prop_Tracts_10_mi_Low_Access=('10 mi Low Access Flag',np.mean),
    Prop_Tracts_20_mi_Low_Access=('1 mi Low Access Flag',np.mean),

    #calucating the total population of a certain race 
    Pop_White = ('White Pop',sum),
    Pop_AA = ('African American Pop',sum),
    Pop_PI_Other =('Pacific Islander Pop',sum),
    Pop_Asian = ('Asian Pop',sum),
    Pop_Native = ('Native Pop',sum),
    Pop_Other_Multiple = ('Other/Mult. Race Pop',sum),
    Pop_Hispanic =('Hispanic Pop',sum),

    Poverty_rate = ('Poverty Rate',np.mean),
    Pop_low_income = ('Low Income Pop',sum)
)

CA.head()
```

In order to make the data visualization of race easier later on, we applied the Python function below to all our counties to determine the proportion of each race per county.

```python
# Calculating proportion of each race in each county in California

def proportion(x):

    prop ={

    'prop_white': (x['Pop_White'].astype('float').sum()) / (x['Population'].sum()),
    'prop_asian': (x['Pop_Asian'].astype('float').sum()) / (x['Population'].sum()),
    'prop_AA': (x['Pop_AA'].astype('float').sum()) / (x['Population'].sum()),
    'prop_PI_Other': (x['Pop_PI_Other'].astype('float').sum()) / (x['Population'].sum()),
    'prop_native': (x['Pop_Native'].astype('float').sum()) / (x['Population'].sum()),
    'prop_other_multiple':(x['Pop_Other_Multiple'].astype('float').sum()) / (x['Population'].sum()),
    'prop_hispanic':(x['Pop_Hispanic'].astype('float').sum()) / (x['Population'].sum()),
    'prop_low_income':(x['Pop_low_income'].astype('float').sum()) / (x['Population'].sum())
    }
    return pd.Series(prop)

prop_race_CA = CA.groupby('County').apply(proportion)

prop_race_CA.head()
```

## Part 2, Section 4: Coefficient and Correlation Analysis
### Introduction

The approach we decided on, based on the fact all our data did not meet all the conditions for the pearson correlation test (the absence of outliers and linearity), was the Kendall Tau-b correlation method.


The Kendall Tau-B correlation method is used to "measure the statistical relationship between two continuous variables." The bigger the magnitude of the coefficient, the more highly connected two variables are.

We'll use the Kendall Tau-B correlation coefficient to examine if there's a link between population of a certain race and Low access and income.

It is important to choose only those columns from the "CA" dataframe which are relevant to our analysis. Because it makes no sense to compare the correlation between, say population of Asians and that of Natives, as they are unrelated and irrelevant to our hypothesis. Below, we make sure to choose only the problems which will help us shed some insight onto what we are trying to find.

```python
kendall_df = CA[['Population','Median_Income', 'Low_Access_and_Income_Pop', 'Pop_White', 'Pop_AA', 'Pop_PI_Other', 'Pop_Asian', 'Pop_Native', 'Pop_Hispanic']]
```

Let us check if the assumptions for Kendall Tau-B correlation test are satisfied before conducting the test. 








### Assumptions for Kendall Tau-B correlation test

1) Each variable should be either ordinal, ratio, or interval measurements

2) It is preferred to have a monotonic relationship between the variables being tested, but this requirement is not a strict one.


The first assumption is met as our data values are numbers only. The second assumption is baseline optional, therefore we can move on with what we have.


### Conducting the Test

We use the inbuilt library in pandas that will help us calculate the Kendall Tau-B coefficient between different columns swiftly.

```python
kendallcorr = kendall_df.corr(method='kendall')
kendallcorr
```

It is difficult to analyze all the numbers obtained above, hence we use a heatmap from seaborn to help us in getting a visual sense on the high correlations and the low ones.

We are choosing to ignore some high correlations between factors which make no sense in the common sense world, such as between the population of 2 races.

```python
sns.heatmap(kendallcorr, 
            xticklabels=kendallcorr.columns,
            yticklabels=kendallcorr.columns,
            cmap='RdBu_r',
            annot=True,
            linewidth=0.5)
```

### Results

From the graph,

1) Strong correlation between Pop_Native and Low_Access_and_Income_Pop. 

2) Weakest correlation between Pop_Native and Median_Income in that column.

3) Weakest correlation with Low_Access_and_Income_Pop with Pop_Asian in the column. 

4) Strong correlation between Pop_White and Median_Income.

### Conclusion: 

1) The Native population is most likely affected the largest by Low access and income factors, and hence low median income.

2) The Asian and White population is least likely to be affected by Low access and income factors. They also have the better correlation with median income when compared to others, suggesting that they have comparatively better access and utilization of resources.

The data set that we will use will be strictly for statistical purposes and will not be used to benefit any person(s) financially or for business purposes.
